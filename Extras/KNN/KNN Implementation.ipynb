# KNN Implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
dfx = pd.read_csv('datax.csv')
dfy = pd.read_csv('datay.csv')
```

```python
X = dfx.values
Y = dfy.values
```

```python
X = X[:,0:]
Y = Y.reshape((-1))
```

```python
print(X.shape)
print(Y.shape)
```

```python
plt.scatter(X[:,0], X[:,1], c=Y)
```

```python
query = np.array([6.5,5])
plt.scatter(X[:,0], X[:,1], c=Y)
plt.scatter(query[0], query[1], color='red')
```

```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

def knn(X,Y,queryPoint,k=5):
    
    vals = []
    m = X.shape[0]
    
    for i in range(m):
        d = dist(queryPoint,X[i])
        vals.append((d,Y[i]))
        
    vals = sorted(vals)
    
    # Nearest K points
    vals = vals[:k]
    vals = np.array(vals)
        
    new_vals = np.unique(vals[:,1],return_counts=True)
    
    index = new_vals[1].argmax()
    pred = new_vals[0][index]
    
    return pred
```

```python
x = knn(X,Y,query)
print(x)
```

```python
x = knn(X,Y,[6,4.5])
print(x)
```
