## Web Scraping

```python
import bs4
import requests
```

```python
url = 'https://www.worldweatheronline.com/bangalore-weather-averages/karnataka/in.aspx'
```

```python
response = requests.get(url)
```

```python
# HTML Parser
soup = bs4.BeautifulSoup(response.content)
```

```python
article_element = soup.findAll('script')
```

```python
scrape = []
for x in article_element[1]:
    scrape.append(str(x))
```

```python
data = scrape[0]
df = data.split()
```

```python
# Max
max_temp = df[10]
```

```python
max_temp = max_temp[43:2852]
```

```python
max_temp = max_temp.split('],[')
```

```python
for i in range(len(max_temp)):
    max_temp[i] = max_temp[i][-2:]
    max_temp[i] = int(max_temp[i])
```

```python
# Min
min_temp = df[12]
```

```python
min_temp = min_temp[43:2852]
```

```python
min_temp = min_temp.split('],[')
```

```python
for i in range(len(min_temp)):
    min_temp[i] = min_temp[i][-2:]
    min_temp[i] = int(min_temp[i])
```

```python
# Avg
avg_temp = df[14]
```

```python
avg_temp = avg_temp[27:2836]
```

```python
avg_temp = avg_temp.split('],[')
```

```python
for i in range(len(avg_temp)):
    avg_temp[i] = avg_temp[i][-2:]
    avg_temp[i] = int(avg_temp[i])
```
